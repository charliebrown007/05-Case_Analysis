De acuerdo con la primera observación, existen valores faltantes en ambos datasets

```python
# Cantidad de valores faltantes en el dataframe
#nfl_data.isnull().sum().sum()
nfl_data.isnull().values.sum()
```

```python
# Valores faltantes por columna
nfl_data.isnull().sum()[nfl_data.isnull().sum() > 0].sort_values(ascending = False)
```

```python
# Valores faltantes por columna
sf_permits.isnull().sum()[sf_permits.isnull().sum() > 0].sort_values(ascending = False)
```

```python
# Valores faltantes en porcentaje para nf1_data: 24,87 %
total_cells = np.product( nfl_data.shape )
total_missing = nfl_data.isnull().values.sum()
(total_missing/total_cells) * 100
```

```python
# Valores faltantes en porcentaje para sf_permits: 26,26 %
total_cells = np.product( sf_permits.shape )
total_missing = sf_permits.isnull().values.sum()
(total_missing/total_cells) * 100
```
